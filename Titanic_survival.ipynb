```python
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/titanic.csv')

# We need to predict the chances of survival on titanic.
```

```python
df.head()
```

**Preparing the Data**

```python
# We do not need all of the columns as input
df01 = df[['Pclass','Sex','Age','Fare','Survived']]
```

```python
df01.shape
```

```python
#in age we have many NaN values
nan_counts = df01.isna().sum()
nan_counts
```

```python
# First we will drop all the na values and check the accuracy.

df02 = df01.dropna(axis=0)
```

```python
nan_counts = df02.isna().sum()
nan_counts
```

```python
df02.shape
```

```python
# We do not need all of the columns as input

input = df02[['Pclass','Sex','Age','Fare']]
target = df02['Survived']
```

```python
input.head()
```

```python
target.head()
```

```python
#Encoding the string values of sex and removing the str column
le_sex = LabelEncoder()

input['sex_num'] = le_sex.fit_transform(input['Sex'])
input.drop('Sex', axis = 'columns',inplace=True)
```

```python
input.head()
```

```python
input.shape
```

```python
# Making the tree
# Choosing the model
model = tree.DecisionTreeClassifier(random_state=42)
```

```python
#Training the model

model.fit(input,target)
```

```python
model.score(input, target)
```

```python
accuracy = model.score(input, target)
print("Accuracy:", accuracy)
```

```python
#plotting the tree
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(100,100))
plot_tree(model, filled=True, feature_names=input.columns)
plt.show()
```
